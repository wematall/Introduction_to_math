Постройте на одном графике две кривые y(x)
для функции двух переменной y(k,x)=cos(k∙x), 
взяв для одной кривой значение k=1, 
а для другой – любое другое k, не равное 1


```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBacked.figure_format = 'svg'
```

```python
# y = np.cos(k*x)

x = np.linspace(-3, 3, 51)

k1 = 1
k2 = 3

y1 = np.cos(k1 * x)
y2 = np.cos(k2 * x)
```

```python
plt.xlabel("Переменная k")
plt.ylabel("Функция f(x)")

plt.title("График y(x)")




plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```
