```python
import numpy as np
import matplotlib.pyplot as plt

from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)

X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)

X, Y = np.meshgrid(X, Y)

Z1 = 2*X + 5*Y # первая плоскость
Z2 = 2*X + 5*Y + 100 # вторая плоскость

ax.plot_surface(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2, color='green')
# ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```

##### 3.1 Напишите код, который будет переводить полярные координаты в декартовы.

```python
import numpy as np

# r - radius; phi - angle

# Полярные в декартовы
def pol2cartesian(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return(x, y)

# декартовы в полярные

def cartesian2pol(x, y):
    r = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(r, phi)

# for exsample
r = 5
phi = 30

print(pol2cartesian(r, phi))
```

### 4. Задание (в программе)

#### 4.1 	Решите систему уравнений:

y = x**2 – 1

exp(x) + x∙(1 – y) = 1

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - y - 1, (np.exp(x) - x*y +x -1))

x1, y1 = fsolve(equations, (1, 1))

print(x1, y1)
```
